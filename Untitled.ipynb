```python
#reading all data we need
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mplt
import matplotlib.pyplot as plt
TRAIN_PATH = "train.csv"
TEST_PATH = 'test.csv'
Full_train = pd.read_csv(TRAIN_PATH,index_col='Id')
Full_test = pd.read_csv(TEST_PATH)
Full_train.dropna(axis=0,subset=['SalePrice'],inplace=True)
y_full = Full_train.SalePrice
X_full = Full_train.drop(axis=1, columns=['SalePrice'])

numeric_col = list((X_full.select_dtypes(exclude='object')).columns)
categorical_col = list((X_full.select_dtypes(include='object')).columns)
low_categorical_col = [col for col in X_full.columns
                  if X_full[col].dtype == 'object' and X_full[col].nunique() <10]
```

```python
#helping function for scoring mae
from sklearn.metrics import mean_absolute_error

def score_mae(model, x_valid,y_valid):
    predict = model.predict(x_valid)
    mae = mean_absolute_error(predict, y_valid)
    return mae
```

```python
#Handle all data we prepared
#drop nan
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(X_train_full[numeric_col], y_full, train_size=0.8, test_size=0.2)

x_drop = x_train.dropna(axis=1)
print(x_drop.shape)
x_drop_valid = x_valid.dropna(axis=1)
print(x_drop_valid.shape)
```

```python
#all models
#Decisiontree models
from sklearn.tree import DecisionTreeRegressor

def find_best_params(x_train, y_train, x_valid, y_valid, max_depth, max_leaf_nodes):
    mae = float('inf')
    best_depth = -1
    best_leaf = -1
    for leaf in max_leaf_nodes:
        for depth in max_depth:
            test_model = DecisionTreeRegressor(random_state=0, max_leaf_nodes=leaf, max_depth=depth)
            test_model.fit(x_train,y_train)
            mae_test = score_mae(model=test_model, x_valid=x_valid, y_valid=y_valid)
            if mae > mae_test:
                mae = mae_test
                best_leaf = leaf
                best_depth = depth
            
    print(mae)
    return best_leaf,best_depth

max_depths = list(range(1,15))
max_leafs = list(range(250,270,1))
#best_leaf, best_depth = find_best_params(x_drop, y_train, x_drop_valid, y_valid,max_depths, max_leafs)
best_depth = 9
best_leaf = 264
#print(f'best depth = {best_depth}\nbest leaf = {best_leaf}')
```

```python
best_param_model = DecisionTreeRegressor(random_state=0, max_leaf_nodes=best_leaf, max_depth=best_depth)
best_param_model.fit(x_drop, y_train)
mae = score_mae(best_param_model,x_drop_valid,y_valid)
print('mae of model with best parameters:',mae)
```

```python
# Another approach, using low categorical columns and pipelines
my_col = numeric_col + low_categorical_col

X_train_full = Full_train[my_col].copy()
X_test_full = Full_test[my_col].copy()
X_train_full.head()
```

```python
# adding new feature
X_train_hasGarage = Full_train[numeric_col+categorical_col].copy()
X_train_hasGarage['HasGarage'] = pd.DataFrame((Full_train['GarageYrBlt'].isnull() == False))
#X_train_full['HasPool'] = pd.DataFrame(X_train_full['HasGarage'])
X_train_hasPool = X_train_hasGarage.copy()
X_train_hasPool['HasPool'] = X_train_full['PoolArea'] > 0
X_train_hasFence = X_train_hasPool.copy()
X_train_hasFence['HasFence'] = X_train_hasPool['Fence'].isnull() == False
print(categorical_col)
```

```python
#preprocessing data for pipelines
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer
from sklearn.preprocessing import OneHotEncoder
# Preprocessing for numerical data
numerical_transform = KNNImputer(n_neighbors=5)
# Preprocessing for categorical data
categorical_transform = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='no_info')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse=False))
])
#Bundle into preprocessing numerical and categorical transforms
preprocessing = ColumnTransformer(transformers=[
    ('num', numerical_transform, numeric_col),
    ('cat', categorical_transform, low_categorical_col) #categorical_col)
])
```
